```python
# prompt: import pnada as pd

import pandas as pd
```

```python
df=pd.read_csv("reslut_int_points.csv")
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure(figsize=(60, 6))  # 创建一个大小为50x6的图形
ax = fig.add_subplot(111, projection='3d')  # 创建一个3D绘图子图

# 绘制三角形表面的三维图
ax.plot_trisurf(df['Points_0'], df['t'], df['u'], cmap='rainbow')

# 设置X、Y、Z轴标签
ax.set_xlabel('x_axis')
ax.set_ylabel('t_axis')
ax.set_zlabel('u_axis')

plt.show()  # 显示图形
```

```python
fig, ax = plt.subplots(figsize=(12,6))
scatter = ax.scatter(df['Points_0'], df['t'], c=df['u'], cmap='rainbow', marker='o', s=100)
ax.set_xlabel('x_axis')
ax.set_ylabel('t_axis')

fig.colorbar(scatter, ax=ax, label='u_axis')
```

```python
fig = plt.figure(figsize=(60, 6))  # 创建一个大小为50x6的图形
ax = fig.add_subplot(111, projection='3d')  # 创建一个3D绘图子图

# 绘制三角形表面的三维图
ax.plot_trisurf(df2['Points_0'], df2['t'], df2['true_u'], cmap='rainbow')

# 设置X、Y、Z轴标签
ax.set_xlabel('x_axis')
ax.set_ylabel('t_axis')
ax.set_zlabel('u_axis')

plt.show()  # 显示图形
```

```python
fig = plt.figure(figsize=(60, 6))  # 创建一个大小为50x6的图形
ax = fig.add_subplot(111, projection='3d')  # 创建一个3D绘图子图

# 绘制三角形表面的三维图
ax.plot_trisurf(df2['Points_0'], df2['t'], df2['Error'], cmap='rainbow')

# 设置X、Y、Z轴标签
ax.set_xlabel('x_axis')
ax.set_ylabel('t_axis')
ax.set_zlabel('u_axis')

plt.show()  # 显示图形
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
norm = plt.Normalize(vmin=df2['Error'].min(), vmax=df2['Error'].max())
ax.scatter(df2['Points_0'], df2['t'], c=df2['Error'], cmap='rainbow', marker='o', s=0.5,norm=norm)

ax.set_xlabel('x_axis')
ax.set_ylabel('t_axis')
fig.colorbar(scatter, ax=ax, label='u_axis')
```
